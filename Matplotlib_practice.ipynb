```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug',
          'Sep','Oct','Nov','Dec']

#Set the y-axis to a list of floats as the total fare in US dollars
#accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82,
         10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y_axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Next, we add the x_axis, y_axis, and yerr=stdev to the errorbar function;
# and to make the error bar look more professional add a "cap" to the error
# bar that's equal to a scalar value. To do this, use the 'capsize=' parameter.
# As it follows:
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars and a capsize using the object-oriented method.

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, label= 'Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()

```

```python
# Create the plot.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add a grid.

plt.grid()

# Add the legend.

plt.legend()
```

```python
# To add major x-axis ticks on a horizontal bart chart,
# we use the xticks() function.
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Create the same bar using the object-oriented method.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Alternatively, you can use 'ax.xaxis.set_ticks' instead of 'ax.set_xticks()'.
# If our y-axis had numerical values, we could change how often every major y-tick
# occurs by using the 'yticks()' function for the MATLAB method and the 
# 'set_yticks()' function for the object-oriented approach.
```

```python
# Add Minor Ticks:
# To add minor ticks to the x-axis on a horizontal bar chart, we use the
# 'set_minor_locator' function with the 'xaxis' class in the object-oriented
# method as it follows: 'x.xaxis.set_minor_locator'.
# Inside this function, we add the 'MultipleLocator' method and pass how often
# we want our minor ticks.

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```

```python
#Creating a Scatter Plot using the MATLAB Method

# 1. Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# 2. Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
len(x_axis)
plt.gca().invert_yaxis()
```

```python
# Create a Bubble Chart using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)

# Adjusting the size by multiplying the data in the y-axis by any value.
# In this case we will multiply each data point in the y-axis by 3.

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
#Then we can use the new y_axis list for the s parameter:
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# We can use the LIST COMPREHENSION to replace for and while loops.
# LIST COMPREHENSION is faster because it is optimized for Python to spot a 
# predictable pattern during a looping.The format for list comprehension is as follows:

# new_list =[expression for item in list if conditional]

# Using the lust comprehension technique to multiply each data point in the 
# y_axis list, we would do the following in the plt.scatter() function:

plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface

# To create a Bubble Chart using the Object-Oriented Interface approach, we'll add the 's' parameter
# just like we did when we used the MATLAB approach:
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Using the Object-Oriented approach, we will make the following changes:

# 1. Change the color of markers to sky blue; and change the size of the 
# markers to 5 times each data point.
# 2. Add a black edge to the color circles; and make the linewidth of the
# circles 2 points.
# 3. Add a legend, x_axis and y_axis labels.
# 4. Switch the axis so that the Fare($) data is on the x_axis, and
# invert the y_axis so that January is at the top.
# Add a limit for the x_axis data.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 5 for i in y_axis], c= "lightblue", linewidths = 2, edgecolors = "black", label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.legend()
ax.set_xlim(0)
plt.show()


```

```python
# Create a Pie Chart Using the MATLAB method

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# To increase the size of the pie chart we will use the following code:
plt.subplots(figsize=(8, 8))

# To add percentages for each month and "explode" the largest percentage
# which in this case is July, the seventh value in the x_axis.
# The "explode" parameter will offset the indicated wedge by a fraction
# of the radius, where "0" is zero distance from the center of the pie,
# and "1" is completely outside the diameter of the pie.

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# We are now going to change the colors of each pie wedge by using the 
# CSS Color in the gallery in the Matplotlib documentation.
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen",
         "greenyellow", "yellow", "orange","gold", "indianred", "tomato",
         "mistyrose"]
explode_values= (0, 0, 0, 0, 0, 0,
                0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```

```python
# Create a pie chart using the Object-Oriented Interface Method

fig, ax= plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Using the Objected-Oriented approach, make the following changes.
# 1. Add a percentage to one decimal place to each wedge.
# 2. Increase the figure to size 8x8
# 3. Explode the two highest percentage months
# 4. Add a shadow
# 5. Add a start angle so that January is at the top
# 6. Reverse the order so that the month order is in a clockwise
# direction
# 7. Add new colors of your own choosing or use the colors from the
# previous chart
fig, ax = plt.subplots(figsize=(8,8))
colors = ["azure", "violet", "lightblue", "palegreen", "yellowgreen",
          "greenyellow", "yellow", "khaki","gold", "indianred",
          "aliceblue","mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()       
```
